## 11.3 重み行列の初期化

### ライブラリのインポート

```python
# ライブラリのインポート
import torch
import torch.nn as nn
```

### PyTorchデフォルトでの初期化

```python
# PyTorchデフォルトでの初期化

# 乱数初期化
torch.manual_seed(123)

# 入力3次元 → 出力2次元の全結合層
# デフォルトではKaiming Uniform（He Uniform)となる
layer = nn.Linear(3, 2)

# 結果確認
print(layer.weight)
print(layer.bias)
```

### PyTorchで重み行列初期化を手動変更

```python
# PyTorchで重み行列初期化を手動変更

# 初期化用ライブラリのインポート
import torch.nn.init as init

torch.manual_seed(123)
layer = nn.Linear(3, 2)

# weightをXavier一様分布
init.xavier_uniform_(layer.weight)

# biasをすべてゼロ
init.zeros_(layer.bias)

# 結果確認
print(layer.weight)
print(layer.bias)
```
